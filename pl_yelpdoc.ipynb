```python
# import 
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
# set up 
driver = webdriver.Chrome("/Users/priyalakra/Desktop/webscrapping_know/webdrivers/chromedriver")

# get url
url = "https://www.yelp.com/"

driver.get("https://www.yelp.com/search?find_desc=Dentists&find_loc=San+Francisco%2C+CA&ns=1")
# wait for 1 sec
time.sleep(1)
```

```python
# accept cookies
# cookie = driver.find_element(by="id", value="onetrust-accept-btn-handler")
# cookie.click()
```

```python
# data we want to scrape
names = driver.find_elements(by="class name", value="css-1422juy")
print(names[1].text)

expertise = driver.find_elements(by="xpath", value=("//div[@class=' border-color--default__09f24__NPAKY']/div/div/div/p[@class='css-1gfe39a']/span[@class=' css-epvm6 display--inline__09f24__c6N_k border-color--default__09f24__NPAKY']"))
print(expertise[1].text.replace("\n", ", "))

ratings = driver.find_elements(by="xpath", value=("//div[@class=' attribute__09f24__hqUj7 display--inline-block__09f24__fEDiJ margin-r1__09f24__rN_ga border-color--default__09f24__NPAKY']/span/div"))
print(ratings[1].get_attribute('aria-label'))

rating_cnt = driver.find_elements(by="xpath", value=("//span[@class='reviewCount__09f24__tnBk4 css-1e4fdj9']"))
print(rating_cnt[1].text)

```

```python
f_names = [i.text for i in names]

#for e in expertise:
  #  print(e.text)
# create an excel file of the extracted data
```

```python
df = pd.DataFrame({"doctor":[i.text for i in names], 
                   "rating":[j.text for j in rating_cnt], 
                   "expertise":[z.text.replace("\n", ", ") for z in expertise],
                   "star":[s.get_attribute('aria-label').replace("star rating", " ") for s in ratings]})
```
